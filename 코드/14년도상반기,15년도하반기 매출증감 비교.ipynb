```python
season = season.drop(['구매평균','구매평균.1','구매평균.2','구매평균.3','구매평균.4','구매평균.5','구매평균.6','구매평균.7'],axis=1)
season
```

```python
a = season[(season['14_1구매경험']=='유') & (season['14_2구매경험']=='유')
       & (season['14_3구매경험']=='유') & (season['14_4구매경험']=='유')
       & (season['15_1구매경험']=='유') & (season['15_2구매경험']=='유')
       & (season['15_3구매경험']=='유')]
a
```

```python
a['총소비금액'] = a['총구매액'] + a['총구매액.1'] + a['총구매액.2'] + a['총구매액.3'] + a['총구매액.4'] + a['총구매액.5'] + a['총구매액.6'] 
a
```

```python
a['구매비율'] = (a['총구매액'] / a['총소비금액']) * 100
a['구매비율.1'] = (a['총구매액.1'] / a['총소비금액']) * 100
a['구매비율.2'] = (a['총구매액.2'] / a['총소비금액']) * 100
a['구매비율.3'] = (a['총구매액.3'] / a['총소비금액']) * 100
a['구매비율.4'] = (a['총구매액.4'] / a['총소비금액']) * 100
a['구매비율.5'] = (a['총구매액.5'] / a['총소비금액']) * 100
a['구매비율.6'] = (a['총구매액.6'] / a['총소비금액']) * 100
a
```

```python
a['분기변동계수'] = np.std(a[['총구매액','총구매액.1','총구매액.2','총구매액.3','총구매액.4','총구매액.5','총구매액.6']],axis=1)/np.mean(a[['총구매액','총구매액.1','총구매액.2','총구매액.3','총구매액.4','총구매액.5','총구매액.6']],axis=1)
a
```

```python
a['객단가'] = a['총구매액'] / a['구매횟수']
a['객단가.1'] = a['총구매액.1'] / a['구매횟수.1']
a['객단가.2'] = a['총구매액.2'] / a['구매횟수.2']
a['객단가.3'] = a['총구매액.3'] / a['구매횟수.3']
a['객단가.4'] = a['총구매액.4'] / a['구매횟수.4']
a['객단가.5'] = a['총구매액.5'] / a['구매횟수.5']
a['객단가.6'] = a['총구매액.6'] / a['구매횟수.6']
a
```

```python
a['객단가변동계수'] = np.std(a[['객단가','객단가.1','객단가.2','객단가.3','객단가.4','객단가.5','객단가.6']],axis=1)/np.mean(a[['객단가','객단가.1','객단가.2','객단가.3','객단가.4','객단가.5','객단가.6']],axis=1)
a
```
